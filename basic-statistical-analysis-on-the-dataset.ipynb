**Analysis of Numerical Features**

```python
print(df['Age'].describe())

print(df['Income'].describe())

sns.histplot(df['Age'], kde=True, bins=30)
plt.title('Age Distribution')
plt.show()

sns.histplot(df['Income'], kde=True, bins=30)
plt.title('Income Distribution')
plt.show()

sns.boxplot(x='Chronic Medical Conditions', y='Age', data=df)
plt.title('Age Distribution by Chronic Medical Conditions')
plt.show()

sns.boxplot(x='Chronic Medical Conditions', y='Income', data=df)
plt.title('Income Distribution by Chronic Medical Conditions')
plt.show()
```

Boxplot: it helps to understand the spread and central tendency of the data.

Comparing age distributions between individuals with and without chronic medical conditions can reveal differences in central tendencies and identify if certain groups have more extreme values.

Boxplot: Provides a summary view in detecting outliers and comparing distributions between categories.


Histplot: Offers a detailed view of the data’s distribution, including its shape and densit

```python
print(df['Age'].describe())

print(df['Income'].describe())

sns.histplot(df['Age'], kde=True, bins=10)
plt.title('Age Distribution')
plt.show()

sns.histplot(df['Income'], kde=True, bins=10)
plt.title('Income Distribution')
plt.show()

sns.boxplot(x='Chronic Medical Conditions', y='Age', data=df)
plt.title('Age Distribution by Chronic Medical Conditions')
plt.show()

sns.boxplot(x='Chronic Medical Conditions', y='Income', data=df)
plt.title('Income Distribution by Chronic Medical Conditions')
plt.show()
```

**Correlation Analysis**

```python
# Correlation matrix
correlation_matrix = df[['Age', 'Income']].corr()
print(correlation_matrix)

# Plot the correlation matrix
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
